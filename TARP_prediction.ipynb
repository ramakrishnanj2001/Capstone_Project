# **Engine Efficiency**

```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = [7.50, 3.50]
plt.rcParams["figure.autolayout"] = True

b1=np.empty(52,dtype="float")
b1[0]=85
for i in range(51):
  b1[i+1]=b1[0]*(0.985**i);

a1=np.empty(52,dtype="float")
a1[0]=59.426
for i in range(51):
  a1[i+1]=(a1[0]*(100-b1[i+1]))/(100-b1[0]);

b2=np.empty(52,dtype="float")
b2[0]=40
for i in range(51):
  b2[i+1]=b2[i]-0.4;

a2=np.empty(52,dtype="float")
a2[0]=121.9
for i in range(51):
  a2[i+1]=(a2[0]*(100-b2[i+1]))/(100-b2[0]);

yr = []
yr = [i for i in range(52)]

xyr = np.array(yr)
yy1 = np.array(a1)
yy2 = np.array(b1)
yy3 = np.array(a2)
yy4 = np.array(b2)

fig1, ax11 = plt.subplots()

ax11.plot(xyr, yy1, '--', color="red")
ax11.set_ylim(0,300)
ax11.set_xlabel("Number of years")
ax11.set_ylabel("Carbon Emission (g/Km)")
ax11.legend(['Carbon Emission (g/Km)'], loc="upper left")

ax12 = ax11.twinx()
ax12.plot(xyr, yy2, '-.', color="blue")
ax12.set_ylim(0,100)
ax12.set_ylabel("Efficiency (%)")
ax12.legend(['Efficiency (%)'], loc="upper right")

fig2, ax21 = plt.subplots()

ax21.plot(xyr, yy3, '--', color="red")
ax21.set_ylim(0,200)
ax21.set_xlabel("Number of years")
ax21.set_ylabel("Carbon Emission (g/Km)")
ax21.legend(['Carbon Emission (g/Km)'], loc="upper left")

ax22 = ax21.twinx()
ax22.plot(xyr, yy4, '-.', color="blue")
ax22.set_ylim(0,100)
ax22.set_ylabel("Efficiency (%)")
ax22.legend(['Efficiency (%)'], loc="upper right")
```
